```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import warnings
warnings.filterwarnings('ignore')

uploaded = files.upload()
```

```python
df = pd.read_csv('large_stock_dataset (1).csv')
df.head()
```

```python
df.info()
print('\n Shape', df.shape)
```

```python
print(df.isnull().sum())
df.fillna(method='ffill', inplace=True)
```

```python
df.describe()
```

```python
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close'], label='Close',marker='.')
plt.title('Closing Price Over 120 Days')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['High'], label='High', color='green')
plt.plot(df['Date'], df['Low'], label='Low', color='red')
plt.title('High and Low Prices Over 120 Days')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Volume'], label='Volume', color='purple')
plt.title('Trading Volume Over 120 Days')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
